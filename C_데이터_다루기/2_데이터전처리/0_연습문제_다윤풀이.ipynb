#### 문제 2

다음과 같은 두 개의 데이터를 하나로 합쳐서 엑셀 파일로 만들고자 한다. 코드를 작성하라.

```python
# 다음과 같은 두 개의 데이터를 하나로 합쳐서 엑셀 파일로 만들고자 한다. 코드를 작성하라.

from pandas import DataFrame, concat

data1 = [
    {'ID': 25, 'READ': 350, 'MATH': 425, 'GRADE': 7},
    {'ID': 27, 'READ': 425, 'MATH': 375, 'GRADE': 8},
    {'ID': 30, 'READ': 475, 'MATH': 485, 'GRADE': 8},
    {'ID': 35, 'READ': 375, 'MATH': 400, 'GRADE': 9}
]
df1 = DataFrame(data1)

data2 = [
    {'ID': 20, 'READ': 250, 'MATH': 300, 'GRADE': 7},
    {'ID': 35, 'READ': 325, 'MATH': 225, 'GRADE': 9},
    {'ID': 37, 'READ': 300, 'MATH': 375, 'GRADE': 9}
]
df2 = DataFrame(data2)

df = concat([df1, df2], ignore_index = True)
df
```

#### (추가내용) Class를 범주형으로 변환

```python
# 생성된 데이터프레임에 대한 각 변수의 데이터 타입 확인
df.dtypes
```

```python
# 데이터프레임의 여러 컬럼을 일괄 변경하여 새로운 데이터프레임으로 생성
df2 = df.astype({'class': 'category'})
df2.dtypes
```

```python
# 특정 컬럼의 타입만 변경 후 원본에 바로 적용 

df['class']=df['class'].astype('category')
df.dtypes
```

```python
# 서열척도로 변경해야 하는 경우

from pandas import CategoricalDtype
df['class'] = df['class'].astype(
    CategoricalDtype(categories = ['A1', 'A2', 'A3', 'A4'], ordered = True)
)
df.dtypes
# 'A4'가 가장 작음?
```

#### 문제 3

다음은 한 도시의 20개 중, 고교를 대상으로 조사한 흡연율 자료이다. 도시와 농촌별에 따른 남여 평균 흡연율을 확인할 수 있는 데이터 피벗테이블로 구성하시오

 https://data.hossam.kr/test/q3.xlsx

```python
from pandas import DataFrame, pivot_table, read_excel

df = read_excel("https://data.hossam.kr/C02/smoke.xlsx")
df1 = pivot_table(df,
            index = '지역',  
            columns = '구성', 
            values = '흡연율',
            aggfunc='mean')
df1.drop('공학', axis = 1)
```

#### 문제 4

다음은 제주도 남녀 대학생의 개인정보 관리방법에 관한 설문조사 결과이다.
자료를 직접 입력하고 남녀별 메모의 종류별로 비율을 확인할 수 있는 데이터를 구성하여라.

https://data.hossam.kr/C02/privacy.xlsx

```python
from pandas import DataFrame, crosstab

df = read_excel("https://data.hossam.kr/C02/privacy.xlsx")
df

df2 = crosstab(index = df['성별'], columns=df['메모종류'],
         rownames=['성별'], colnames=['메모종류'],
         normalize=True)
df2
```
