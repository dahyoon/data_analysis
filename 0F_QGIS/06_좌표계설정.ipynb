# 좌표계

- : 지리적 객체를 위치시키는데 사용되는 좌표기반체계
- 두 가지 종류: `지리좌표계(GCS)`와 `투영좌표계(PCS)`로 구분

## I. 좌표계 종류
### 1) `지리좌표계(GCS: Geographic Coordinate System)`
- 지구상의 위치를 정의하는데 3차원의 구면을 이용하는 좌표계
- 한 지점은 `경도(longitude, x축)와 위도(latitude, y축)` 값으로 지정
- 위도와 경도는 지구중심에서 지표면상의 지점까지 측정된 각으로 흔히 도(degree)로 표시
- QGIS 상에서 지도를 이동할 경우 축적이 함께 변한다

### 2) 투영좌표계(PCS: Projected Coordinate System)
- 지리좌표를 평면상에 투영한 좌표계
- 하나의 타원체를 투영하는데 `다양한 방식의 투영법`들이 있음
- QGIS 상에서 지도를 이동하더라도 축적이 고정되어있다

### 대한민국에서 자주 사용되는 좌표계
- 경위도 (ESPG: 4326)
- KLIS 중부 혹은 Bessel 보정된 중부원점 (ESPG: 5174)
- 네이버 (ESPG: 5179)
- 다음 (ESPG: 5181)
- 구글 (ESPG: 3857)
- 국토지리정보원 중부 (ESPG: 5186)
- 국토지리정보원 동부 (ESPG: 5187)
- UTM52N (ESPG: 32652)

    - 다른 좌표계를 갖고 있는 파일들의 좌표계를 통일시켜야 할 경우가 있어서 좌표계를 맞추는 방법을 알아야 한다
- 저장자

    |포멧|설명|
    |--|--|
    |`*.dbf`|각 도형의 속성 정보가 저장되어 있는 파일|
    |`*.prj`|좌표정보가 저장되어 있는 파일|
    |`*.shp`|shape format. 도형파일(백터형식)|
    |`*.shx`|shape index format. 도형의 위치, 방향 정보가 저장되어 있는 파일|

### 1) 좌표계 설정하기
#### 0. 프로잭트의 기본 좌표계
- 프로잭트의 좌표계는 처음 레이어의 좌표계를 따라간다
- 처음 어떤 레이어를 불러오느냐가 그 프로잭트의 기본 좌표계로 자동설정 된다
- 지도 파일을 다운로드 받을 때 그 파일에 .prj 파일(좌표계)이 있는지, 사이트에 어떤 좌표계를 따르는지 명시되어 있는지를 확인해야 된다
    - 좌표계를 바르게 잡기 위하여
#### 1. 프로잭트 내 레이어 좌표계 설정
- shape파일의 건물 레이어 추가하기
    - shape 파일 (/Users/dayoonz/Desktop/data_analysis_local/F_GIS/Lab06_좌표계설정/수치지도/b0010000_polygon.shp) 더블클릭
    - 해당 파일 (b0010000_polygon) 우클릭 > Layer CRS > Set Layer CRS > Predefined CRS > Filter: 원하는 좌표계 (ESPG:5186) 입력 > 선택 후 OK
#### 2. 좌표계 없는 파일들의 좌표계 프로잭트 내 자동 설정
- Settings > Option > CRS and Transforms > CRS Handling > CRS for Layers > Default CRS for layers: EPSG:5186 (한국 데이터들은 이 좌표계를 많이 사용함)
    - 밑에 체크들 (Leave as an unknown CRS (take no action), Prompt for CRS, Use project CRS, Use default layer CRS)은 사용상 편의에 따라 설정하면 됨
    - 보통 Use default layer CRS 해줌
#### 3. 좌표계 설정을 컴퓨터에 저장 (.prj 파일 생성) - 좌표계 없는 파일들 혹은 좌표계를 바꾸고자 하는 파일들
- Processing Toolbox (Define Shapefile projection 검색해도 됨) > Vector General > Define Shapefile projection > Parameters > Input Shapefile: 좌표계를 설정하고자 하는 shapefile (b0010000_polygon [EPSG:5186]), CRS: 설정하고자 하는 좌표계 (EPSG:5186 = Korea 2000 / Central Belt 2010) > Run
    - 해당 폴더에 보면 .prj 파일이 생성되어있을 것이다
- 레이어에 좌표계를 설정하여 Export > Save feature As ... > 여기서 좌표계 설정하면 .prj 파일도 같이 생성된다
- 건물정보 저장하기
    - 해당 shape파일 레이어로 추가 (/Users/dayoonz/Desktop/data_analysis_local/F_GIS/Lab06_좌표계설정/건물통합정보_서울/Archive/F_FAC_BUILDING_11_202309.shp)
    - 좌표계 설정 (해당 레이어 우클릭 > Layer CRS > Set Layer CRS > Predefined CRS > Filter: 원하는 좌표계 (ESPG:5174) 입력 > 선택 후 OK)

### 2) 좌표값 확인
- 프로젝트에 두 레이어 추가 (/Users/dayoonz/Desktop/data_analysis_local/F_GIS/Lab06_좌표계설정/crs/4326.shp), (/Users/dayoonz/Desktop/data_analysis_local/F_GIS/Lab06_좌표계설정/crs/23030.shp)
- 해당 레이어 선택 (둘 중 아무거나, 4326) > Vector > Geometry Tools > Add Geometry Attributes > Parameters > Input layer: 4326 [EPSG:4326], Calculate using: Layer CRS(해당 레이어의 좌표계를 그대로 사용하겠다는 의미), Added geom info: 건들지 마 [Create temporary layer], Open output file after running algorithm: chekced > Run
    - Added geom info라는 레이어가 생성되었을 것이다
        - 이 레이어의 속성 테이블을 확인하면 (우클릭 > Open Attribute Table) 2개의 컬럼 (해당 좌표계의 x축, y축 값들)이 추가 됨
- 다른 레이어 선택 (23030) > Vector > Geometry Tools > Add Geometry Attributes > Parameters > Input layer: 23030 [EPSG:23030], Calculate using: Layer CRS(해당 레이어의 좌표계를 그대로 사용하겠다는 의미), Added geom info: 건들지 마 [Create temporary layer], Open output file after running algorithm: chekced > Run
    - Added geom info라는 레이어가 생성되었을 것이다
        - 이 레이어의 속성 테이블을 확인하면 (우클릭 > Open Attribute Table) 2개의 컬럼 (해당 좌표계의 x축, y축 값들)이 추가 됨
> 좌표값을 확인하면 서로 다르다 - 서로 다른 좌표계를 사용하고 있기 때문
>
> 이럴 때, 지도상 보이는 것은 같을지라도 실제 값/위치는 다를 수 있다. 그러므로 좌표계를 통일시켜줘야 된다
### 3) 좌표계 바꾸기 - 바꿔서 저장
- 지도상에는 같은 위치에 있지만 서로 좌표계가 다르게 설정되어 있는 경우, 아래의 방법들을 통하여 좌표계를 통일해주야 나중에 정밀작업해야할 때 
#### 방법 1. 
- Processing Toolbox (Reproject layer 검색) > Vector General > Reproject layer > Parameter > Input layer: 바꾸고자 하는 레이어 (23030 [EPSGL23030]), Target CRS: 원하는 좌표계 설정, Reprojected: 저장할 경로랑 이름 (23030_to_4326.shp)설정, `shape file`로 저장!
#### 방법 2. 
- 해당 레이어 우클릭 > Export > Save Feature As ... > CRS: 원하는 좌표계 설정 > OK
> 이렇게 좌표계를 바꿔서 저장해 놓은 후, 좌표값을 다시 확인해보면 (Vector > Geometry Tools > Add Geometry Attributes ... 위에 확인) 같은 좌표값인 것을 확인할 수 있다 (4326과 23030_to_4326)
### 3) 없는 좌표계 추가하기

+towgs84=-145.907,505.034,685.756,-1.162,2.347,1.592,6.342 +units=m
+no_defs) > Validate 클릭 > OK 클릭
