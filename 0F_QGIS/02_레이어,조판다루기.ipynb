### 5) 시각화 심볼 설정
- 시도별 다른 색상
    - 시도(레이어) > Properties > Symbology > 첫번째 박스: Categorized, Value: SID_KOR_NM(시도한글명), Color ramp: 원하는 색상 지정 > Classify 버튼 누르기 > Apply
        - 시도별로 다른 색상이 입혀진 지도가 보여진다
- 인구수에 따른 다른 색상
    - 시도(레이어) > Properties > Symbology > 첫번째 박스: Graduated, Value: POP2015, Color ramp: 원하는 색상 지정 > Mode: Equal Count(Quantile), Classes: 5 (원하는만큼 나누기) > Classify 버튼 누르기 > Apply
        - Color ramp 밑에 Classes > Legend 값들을 더블클릭해서 값을 바꾸면 지도에 표시되는 범례를 변경할 수 있다(수치값은 안 바뀜)

### 6) 보고서에 사용할 수 있는 파일로 만들기 (조판 Layout)
- 새 조판 Layout 만들기
    - Project > New Print Layout > Add Item (메뉴바) > Add Map 누른 뒤 드래스하면 작업하던 지도가 그려짐
- 작업하던 조판  Layout 불러오기
    - Project > Layout Manager
- 축적
    - Add Item > Add Scale Bar
- 범례
    - Add Item > Add Legend
    - 불필요한 항목들 제거하고 싶을 때 오른쪽 하단 Item Properties에서 Legend Items 아래 Auto Update 체크 풀기
- 문자 
    - Add Item > Add Label
- North Arrow
    - Add Item > Add North Arrow
- 해상도 
    - Layout > Export Settings > Export Resolution:   
        - 72 dpi는 모니터 기본값  
        - 150 dpi - 일반 모니터라 해도 최소한 150 dpi 넣어주기 (요즘 모니터 해상도가 좋아서 72로 잡으면 흐릿한 경우가 많다)
        - 300 dpi는 인쇄물/프린트 용
        - 600 dpi는 포스터 사이즈 정도 용
- 파일 저장하기
    - Layout > Export as (원하는 파일의 형태)
