### 문제 1
- 다음 데이터는 조선시대 역대 왕의 수명을 조사한 데이터이다: https://data.hossam.kr/D02/kings_life.xlsx
- 수명에 대해 기술통계량을 구하고 상자그림을 그려라
- 최종적으로 분석 결과를 토대로 하여 알 수 있는 사실을 하나 이상 제시하라.

```python
import sys
from pandas import read_excel
from matplotlib import pyplot as plt

data = read_excel('https://data.hossam.kr/D02/kings_life.xlsx', index_col="왕")
data

plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (7, 4)
plt.rcParams["axes.unicode_minus"] = False

print(data.describe())

plt.figure()
plt.title('조선시대 왕들의 수명')
plt.grid()
data.boxplot()
plt.show()
plt.close()
print("Insight: 조선시대 왕들의 평균 수명은 45세 정도 된다.")
print("Insight: 총 27명의 왕에 대한 수명은 16-82 사이의 범위를 갖고 있으며\n평균 수명은 46세, 중앙값은 45세, 1사분위 수는 33, 3사분위 수는 56이다.\n상자그림을 통해 이상치는 없음을 확인할 수 있다.")
```

### 문제 2

- 외국기업과 내국기업별로 기술통계량을 상자그림을 그려서 비교/분석하라

```python
import sys
import seaborn as sb
from pandas import read_excel
from matplotlib import pyplot as plt

plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (5, 2)
plt.rcParams["axes.unicode_minus"] = False

data = read_excel('https://data.hossam.kr/D02/stock.xlsx')
data

print(data.describe())
dataF = data.query("구분 == 'F'")
dataF.reset_index(drop = True, inplace=True)
dataK = data.query("구분 == 'K'")
dataK.reset_index(drop = True, inplace=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))
fig.suptitle('내국기업/외국기업 주가 비교', fontsize = 36, color = '#000000')
sb.boxplot(y='주가', data=dataF, ax = ax1, color = 'lightblue')
ax1.set_xlabel('외국기업', fontsize = 15)
ax1.set_ylim(100, 240)
sb.boxplot(y='주가', data=dataK, ax = ax2, color = 'pink')
ax2.set_xlabel('내국기업', fontsize = 15)
ax2.set_ylim(100, 240)
plt.show()
plt.close()
print("Insight: 외국 기업의 주가보다 내국 기업의 주가의 범위가 더 넓다.")
print('Insight:\n1. 주가의 변동 폭은 F보다 K가 더 크다.\n2. K는 폭락 지점이 보이고, F는 상승 지점이 보인다.')
```

### 문제 3
- 포도농장에서 수집한 비료 종류에 따른 수확량 자료이다: https://data.hossam.kr/D02/grape.xlsx
- 기술통계량을 구하고 상자그림을 그려라.
- 최종적으로 분석 결과를 토대로 하여 알 수 있는 사실을 하나 이상 제시하라.

```python
import sys
import seaborn as sb
from pandas import read_excel
from matplotlib import pyplot as plt

plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (5, 2)
plt.rcParams["axes.unicode_minus"] = False

data = read_excel('https://data.hossam.kr/D02/grape.xlsx')

print(data.describe())

plt.figure()
plt.grid() 
sb.boxplot(x='비료종류',y='수확량',data=data)
plt.title('비료 종류 별 수확량',fontsize=17)
plt.show()
plt.close()
print("Insight: B 비료종류의 수확량의 범위가 A 비료종류의 수확량의 범위보다 넓지만, 수확량 자체는 더 낮다.")
print("Insight: 평균값을 비교했을 떄 A 비료를 사용하는 것이 B를 사용하는 것보다 더 많은 수확량이 보인다.\n수확량 분포는 B비료를 사용하는 것이 더 넓게 분포된 것으로 보아 A를 사용하는 것이 더 안정적으로 보인다.")
```

```python
import sys
import seaborn as sb
from pandas import read_excel
from matplotlib import pyplot as plt

plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (5, 2)
plt.rcParams["axes.unicode_minus"] = False

data = read_excel('https://data.hossam.kr/D02/grape.xlsx')

print(data.describe())

dataA = data.query("비료종류 == 'A'")
dataA.set_index('비료종류', inplace = True)  
dataB = data.query("비료종류 == 'B'")
dataB.set_index('비료종류', inplace = True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 4))
sb.boxplot(data = dataA, y = '수확량', ax = ax1, color = 'lightcoral')
ax1.set_title('A비료', color = 'lightcoral')
ax1.set_ylim(10, 40)
sb.boxplot(data = dataB, y = '수확량', ax = ax2, color = 'lightskyblue')
ax2.set_title('B비료', color = 'lightskyblue')
ax2.set_ylim(10, 40)
plt.show()
plt.close()
print("Insight: B 비료종류의 수확량의 범위가 A 비료종류의 수확량의 범위보다 넓지만, 수확량 자체는 더 낮다.")
print("Insight: 평균값을 비교했을 떄 A 비료를 사용하는 것이 B를 사용하는 것보다 더 많은 수확량이 보인다.\n수확량 분포는 B비료를 사용하는 것이 더 넓게 분포된 것으로 보아 A를 사용하는 것이 더 안정적으로 보인다.")
```

### 문제 5
- 다음은 통계학 전공(STAT) 학생과 전산학 전공(COMP) 학생의 통계학 과목에 대한 중간고사 성적과 기말시험 성적을 나타낸 표이다: https://data.hossam.kr/D02/stat_comp_grade.xlsx

1. 전공에서 STAT를 1로, COMP를 2로 변환한 레이블을 적용하라.
2. 중간고사 및 기말고사 성적에 대한 각종 기술통계량을 구하고, 분석 결과를 토대로 하여 알 수 있는 사실을 하나 이상 제시하라.
3. 중간고사 및 기말고사 성적에 대한 히스토그램을 그리고, 분석 결과를 토대로 하여 알 수 있는 사실을 하나 이상 제시하라.
4. 전공별로 중간고사 및 기말고사 성적에 대한 히스토그램을 그리고, (3)번의 결과와 비교 하라.

```python
import sys
import seaborn as sb
import numpy as np
from pandas import read_excel
from matplotlib import pyplot as plt

plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (5, 2)
plt.rcParams["axes.unicode_minus"] = False

data = read_excel('https://data.hossam.kr/D02/stat_comp_grade.xlsx')

data['전공'] = np.where(data['전공'] == 'STAT', 1, 2)
data['전공'] = data['전공'].astype('category')

data1 = data.query("전공 == 1").reset_index().drop('index',axis = 'columns')
data2 = data.query("전공 == 2").reset_index().drop('index',axis = 'columns')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 4))
fig.subplots_adjust(wspace = 0.3)
fig.suptitle('1과 2의 중간/기말고사 성적 비교', fontsize = 15, color = '#000000')
data1.boxplot(['중간고사', '기말고사'], ax = ax1)
ax1.title.set_text('1: STAT')
ax1.title.set_fontsize(7)
data2.boxplot(['중간고사', '기말고사'], ax = ax2)
ax2.title.set_text('2: COMP')
ax2.title.set_fontsize(7)
plt.show()
plt.close()
print("Insight: .")

# hist1, bins1 = np.histogram(data['기말고사'], bins = 5)
# bins2 = np.round(bins2, 1)
# bins2
```
