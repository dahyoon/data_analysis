# 데이터프레임 시각화

데이터프레임이 내부적으로 pyplot 객체를 초기화 하기 때문에 plt.figure()를 호출할 필요가 없다.

### 1. 작업준비

#### 1) 패키지 참조

```python
from pandas import read_excel
from matplotlib import pyplot as plt
```

#### 2) 데이터 가져오기

년도별 교통사고 발생건수, 사망자수, 부상자수 데이터

> 데이터 출처: KOSIS 국가통계포털

```python
df = read_excel("https://data.hossam.kr/D01/traffic_acc_year.xlsx", index_col='년도')
df
```

#### 3) 그래프 전역 설정

```python
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 10
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.unicode_minus'] = False
```

### 2. Line Plot

#### 1) 특정 컬럼에 대한 시각화
- dataframe명['컬럼명'].plot()
- 기본적으로 DataFrame의 컬럼 이름이 label이 되지만 .plot() 메서드에 label 파라미터를 설정하여 다른 이름으로 변경 가능

```python
df['발생건수'].plot(label = '교통사고 발생건수')
plt.title('교통사고 발생수')
plt.grid()
plt.legend()
plt.show()
plt.close()
```

#### 2) 2개 이상 컬럼에 대한 시각화
- dataframe명.filter(['컬럼1', '컬럼2']).plot()

```python
df.filter(['사망자수', '부상자수']).plot(rot =45)
plt.title('교통사고 사망자수 & 부상자수')
plt.grid()
plt.legend()
plt.show()
plt.close()
```

#### 3) 전체 컬럼에 대한 시각화
- dataframe명.plot()

```python
df.plot()
plt.title('년도별 교통사고')
plt.grid()
plt.legend()
plt.ylabel('교통사고')

plt.show()
plt.close()
```

### 3. Bar Plot

#### 1) 특정 컬럼에 대한 시각화
- dataframe명['컬럼명'].plot.bar()
- () 안에 스타일 적용 가능:
    - rot = 45 --> x ticks 45도 기울어짐

```python
df['발생건수'].plot.bar(rot =45)
plt.title('년도별 교통사고 변화')
plt.grid()
plt.ylim(200000, 240000)
plt.show()
plt.close()
```

#### 2) 2개 이상 컬럼에 대한 시각화
- dataframe명.filter(['컬럼명1', '컬럼명2']).plot.bar()

```python
df.filter(['발생건수', '부상자수']).plot.bar(rot = 45)
plt.title('년도별 교통사고 변화')
plt.grid()
plt.legend()
plt.ylim(150000, 370000)
plt.show()
plt.close()
```

#### 3) 전체 컬럼에 대한 시각화
- dataframe명.plot.bar()

```python
df.plot.bar(rot = 45)
plt.title('년도별 교통사고 변화')
plt.grid()
plt.legend()
plt.show()
plt.close()
```

### 4. 산점도 그래프
- dataframe명.plot.scatter(x = '컬럼명1', y = '컬럼명2' [, 스타일적용])

```python
df.plot.scatter(x = '발생건수', y = '부상자수', color = 'pink', rot =45)
plt.grid()
plt.show()
plt.close()
```

### 5. Pie 그래프
- dataframe명['컬럼명'].plot.pie(autopct = '각 조각 안에 들어갈 컬럼에 대한 내용')
- dataframe의 index가 파이 그래프의 각 조각이 됨

```python
df['사망자수'].plot.pie(autopct = '%0.2f%%') # 소수점 1자리수까지 표시 + '%' 표시
plt.show()
plt.close()
```
