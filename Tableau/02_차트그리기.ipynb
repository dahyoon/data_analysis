# 태블로에서 차트 그리기
- 측정값과 차원
- 목록들을 시각화하고 싶을 때 drag and drop, 또는 더블 클릭
- 목록 다중 선택하고 싶을 때 ctrl/command 누른 상태에서 목록들 선택 시 다중 선택 가능 (후 오른쪽 상단의 `표현 방식` 클릭 후 원하는 그래프 선택 시 원하는 그래프 종류로 데이터 시각화 가능)
- 태블로는 항상 측정값을 화면 위로 시각화할 시 각 각의 데이터를 보여주지 않고 `하나의 큰 덩어리`로 `집계`해서 표현된다 (예, sales 더블 클릭 시 합계라는 하나의 큰 덩어리가 표현됨)
- 비율 표시하고 싶을 때 해당 목록 우클릭 > `퀵 테이블 계산` > `구성 비율` 체크

## I. 바 차트, 라인 차트, 영역 차트
### [1] 바 차트 Bar
- Super Store 데이터 불러와서 Region별로 Sales 보여주는 바 차트 만들기
    - 열/행 선반에 Region과 Sales를 추가하면 태블로는 저절로 바 차트를 생성 (바 차트가 가장 적합한 표현방식이라고 태블로가 자체 판단 후 선택했기 때문이다)
    - 표현방식 바꾸기: '마크' 박스 안에 있는 '자동' 클릭 후 원하는 표현방식 선택
- 눈에 잘 보이는 바 차트 만들기
    1) `불필요한 필드 레이블 숨기기`
        - Region이라는 레이블을 숨기고 싶은 경우: 레이블 우클릭 > 열에 대한 필드 레이블 숨기기 클릭
    2) `축 머리글 표시 해제 후 레이블 직접 붙이기`
        - Sales 값들을 나타내는 y축 머리글을 숨기고 싶은 경우: y축 레이블 칸 우클릭 > 머리글 표시 클릭 해제
        - 바 위에 레이블 표시하고 싶은 경우: 위 아이콘들 중 T 써져있는 '마크 레이블 표시' 클릭
    3) `불필요한 참조선 없애기`
        - 배경에 있는 격조선(grid) 없애고 싶은 경우: 시트 배경 우클릭 > 서식... > '라인' 아이콘 클릭 > 행 > 격자선: 없음 선택
        - y축에 있는 축 눈금자 없애고 싶은 경우: 시트 배경 우클릭 > 서식... > '라인' 아이콘 클릭 > 시트 > 축 눈금자: 없음 선택
    4) `0이 되는 기준선 표시하기`
        - 웬만하면 이 선을 표시하는 것 권장 - 이 선이 없으면 바 차트가 허공에 둥둥 떠다니는 느낌을 줌
        - 0이 되는 기준선 표시하고 싶은 경우: 시트 배경 우클릭 > 서식... > 시트 > 영(0) 기준선: 원하는 스타일 선택
    5) `(필요에 따라) 불투명도 조절하기` - 눈의 피로감 조절
        - '마크' 박스 안에 있는 '색상' 박스 클릭 > 불투명도 > 원하는 만큼 불투명하게 설정 가능
    6) `(필요에 따라) 짙은 색상의 테두리 활용` - 더 선명하게 보이기 위해
        - '마크' 박스 안에 있는 '색상' 박스 클릭 > 효과 > 테두리: 원하는 색상 설정 가능
### [2] 라인 차트 Line
- 전체적인 흐름/트렌드를 보기에 적합한 차트 타입
- Super Store 데이터 불러와서 Order Date별로 Sales 보여주는 바 차트 만들기
    - 열/행 선반에 Order Date와 Sales를 추가하면 태블로는 저절로 라인 차트를 생성 (라인 차트가 가장 적합한 표현방식이라고 태블로가 자체 판단 후 선택했기 때문이다)
    - 날짜 필드 (년(Order Date)) 앞의 + 기호: 날짜 데이터의 계층 구조를 의미
        - 날짜 데이터의 계층 구조: 연 - (분기) - 월 - 일 - 시 - 분 - 초
        - + 클릭 시 현재 나와있는 '년' 다음인 '분기' 별 데이터로 나옴
        - 한 번 더 + 클릭 시 다음인 '월' 별 데이터로 나옴, 다음은 '일' 별 데이터로 나옴
- 연도 + 월 별 데이터로 보고 싶은 경우: 년(Order Date)와 월(Order Date)를 열에 놓기

    2) `머리글 표시 해제하여 축 숨기기/표시하기`

    3) `불필요한 참조선 없애기`

    4) `필요한 항목에 적절히 레이블링 하기`
        - 지점 위에 레이블 표시하고 싶은 경우: 위 아이콘들 중 T 써져있는 '마크 레이블 표시' 클릭 (이럴 경우 월 별 레이블이 표시됨)

    5) `(필요에 따라) 라인의 두께, 색상, 마커를 적절히 조정`
        - '마크' 박스 > 
            - 색상 >
                - `색상 설정 가능`
                - 효과 > 마커: 선 위에 동그라미 점들로 `마커 표시 가능`
            - 크기 > `두께 설정 가능`

### [7] 히스토그램
- 분포를 나타낼 때 사용
- 도수분포: 특정 항목 또는 특정 범위에 속하는 `빈도수를 나타낸 표`
- 즉, 각 `구간`에 담길 데이터들을 나누는 것
- 주요 개념:<br>

| 용어 | 의미 |
|---|---|
| 계급<br/>(Class) | 자료가 취하는 전체 범위를 몇 개의 소집단(범주,구간)으로 나눈 것
| 도수/빈도<br/>(Frequency) | 각 계급에 속하는 자료의 수 |
| 상대도수<br/>(Relative Frequency) | 도수를 전체 자료의 수, 즉 전체 도수로 나눈 비율(Proportion,Probability)<br/>모든 상대도수의 합 = 1 |
| 도수분포표<br/>(Frequency Table) | 데이터의 대략적인 분포 형태, 중심위치, 산포 등을 파악하기 위한 데이터정리 방법<br/>계급, 도수, 상대도수, 누적도수 등으로 구성 |

- `구간차원 만들기`: 왼쪽 메뉴에서 `해당 필드 우클릭` > `만들기` > `구간차원` ... > `구간차원 크기`에 원하는 값 입력<br>
&emsp;&emsp;<img src="image17_histogram.png" width="50%">

- `비율 표시`: `해당 필드 우클릭` > `퀵 테이블 계산` > `구성 비율` 클릭

### [9] 기타: 트리맵(사각형), 버블맵(원형)
#### 트리맵
- 1개 이상의 차원, 1개 또는 2개의 측정값 필요
- 왼쪽 테이블에서 Category, Sub-Category, Profit, Sales를 ctrl/command 누른 상태에서 선택 후 표현 방식에서 트리맵 쿨릭 시<br>
&emsp;&emsp;<img src="image24_tree.png" width="50%"><br>
- 마크 아래 Sales를 크기에 drag and drop, Profit을 색상에 drag and drop 할 시<br>
&emsp;&emsp;<img src="image25_correct.png" width="50%"><br>
- 트리맵 설명:
    - 왼쪽 상단이 technology와 관련된 부분, 오른쪽이 Office Supplies, 왼쪽 하단이 Furniture와 관련된 부분
    - 각각의 카테고리 내에서 sub category로 더 상세하게 분류되어 있음
    - Furniture의 Tables, Bookcases 및 Office Supplies의 Supplies가 매출에 문제가 되고 있다는 것을 확인
- 트리맵의 단점:
    - 넓이로 상대적인 크기를 인지하는 것이 직관적으로 다가오지 않음
    - 공간 확용에서의 비효율성 &rightarrow; 공간을 너무 많이 차지함, 각 박스 안에 comment를 달고 싶을 때 공간이 넓은 곳에는 내용 작성이 수월하지만 작은 곳에는 내용이 잘림

#### 버블맵
&emsp;&emsp;<img src="image26_bubble.png" width="50%"><br>
- 버블맵 설명:
    - Sales 합계 크기: 원의 색깔 &rightarrow; 색이 진할수록 Sales가 큼
    - Profit 합계의 크기: 원의 크기 &rightarrow; 원이 클수록 Profit이 큼
- 버블맵의 단점:
    - 트리맵과 같이 넓이로 상대적인 크기를 인지하는 것이 직관적으로 다가오지 않음
    - 그러므로 실무에선 많이 활용되지 않음
