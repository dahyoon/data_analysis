# 태블로 주요 기능 익히기
- 데이터: Super Store의 Order 테이블 (xlsx 파일 사용, 출처는 Tableau 웹사이트)
## 1. 필터
- 대용량의 데이터에 추출 필터 및 데이터 원본 필터 적절히 사용하면 화면의 퍼포먼스를 개선하는 데 좋음
- 태블로는 다양한 필터를 제공하나 각각의 필터들이 작동하는 우선순위가 다르기에 화면에서 어떤 계산을 사용하는지 (예 - 집계 계산, LOD? 계산, 테이블 계산을 사용하는지 등)를 고려하여 내가 원하는 화면/결과를 만들기 위해 신중하게 선택하여 사용해야함
### 1) 추출 필터
- 데이터 원본 소스에서 데이터의 일부분만 추출하고 싶을 때 사용
- Tableau Desktop에서만 가능? (Tableau Public에는 이 기능 없음)
- 오른쪽 상단에 `연결` 아래 라이브랑 추출 중 `추출` 선택 후 오른쪽으로 `편집` 클릭
&emsp;<img src="image33_추출필터.png" width=50%>
- 가운데 필터 박스에서 `추가` 클릭 후 원하는 필터 (예, 2016 ~ 2020 데이터 중 2018년도 데이터만 추출하도록 필터 걸기) 선택 후 `확인` 클릭
- 하단에 `시트1` 클릭 시 해당 데이터를 어느 경로에 저장할 지 뜸
- 노트: 현재의 통합문서가 TWBX일 경우에는 저장 경로 창이 나타나지 않음
### 2) 데이터 원본 필터
- 오른쪽 상단에 `필터` 아래 `추가` 클릭<br>
&emsp;<img src="image34_data.png" width=50%>
- 작업을 위한 데이터 중 일부만을 워크 스페이스로 불러올 때 사용
- `추가` 클릭 후 원하는 필터 (예, Category 안에 Technology 필터 걸기) 선택 후 `확인` 클릭)
- 해당 데이터만 나올 것임

### 4) 차원 필터
- 차원 아래에 있는 필드를 필터로 적용하는 것
- 우선 그래프 그리기 (예, 행: Sales, 열: Region - 지역 별 판매액)
- Sub-Category (차원)를 필터 박스에 drag and drop > `일반` 아래 `목록에서 선택` 아래 체크박스 전체 선택 후 `확인`
- 더 쉽게 필터 적용하고 싶은 경우 `필터` 박스 안 `Sub-Catogory`화살표 표시 아래 `필터 표시` 선택 시 오른쪽 바에 필터 목록이 표시됨<br>
&emsp;<img src="image35_filter.png" width=50%>
- 단일값/다중값 선택 가능: 단일값으로 선택 시 라디오버튼/드롭다운/슬라이더로 필터가 바뀌면서 필터 적용됨
&emsp;<img src="image36_filter.png" width=50%>

### 5) 측정값 필터
- 측정값 아래에 있는 필드를 필터로 적용하는 것
- 우선 그래프 그리기 (예, 행: Sales, 열: Region - 지역 별 판매액)
- Sales (측정값)를 필터 박스에 drag and drop > 창이 뜰 것: `집계된 측정값`으로 필터링 할 것인가, RAW 레벨에서 필터링 할 것인가 선택해야 됨
    - 집계된 측정값으로 필터링 예) `#합계` 선택 후 `다음>` 클릭 시
        - 합계([Sales])에 대해 필터링 범위를 지정 가능<br>
    &emsp;<img src="image37_filter.png" width=50%>
    - 데이터 셋의 RAW 레벨에서 필터링 예) 맨 위 `#모든 값` 선택 후 `다음>` 클릭 시
        - 아까와는 다른 범위 값 확인 가능 (더 넓은 범위)
        - `RAW 데이터에 대한 값 필터링` (합계를 거친 데이터에 대한 값과는 다름)
    &emsp;<img src="image38_filter.png" width=50%>

### 7) 숨기기
- 우선 그래프 그리기 (예, 행: Sales, 열: Region - 지역 별 판매액)

## 2. 이중축 (+ Combined Axis Chart)
- 3가지 용도: 1) 이중축의 의미에 충실한 기본 사용법 2) 강조 포인트 제공 3) 디자인 효과 또는 컨텍스트 부여

### 1) 이중축 기본 사용법 
- 우선 그래프 만들기: Sales와 Profit을 행에, Sub Category를 열어 넣어 두 개의 그래프 생성
<img src="image51_dualaxis.png" width=50%>
- 이중축 만드는 방법 2가지
    - 1 - 두번째 측정값을 화면 오른쪽으로 점선이 보일 때까지 끌기 &emsp;<img src="image48_dualaxis.png" width=50%>
    - 2 - 두번째 측정값 우클릭 > `이중 축` 클릭 &emsp;<img src="image50_dualaxis.png" width=50%>
        - 결과: <img src="image49_result.png" width=50%>
- 이중축 생성 시 각 측정값에 따른 마크 카드가 생성됨
    - 마크 카드의 속성을 바꿔서 시각화 조절 가능: 색상, 투명도, 크기 등
    &emsp;&emsp;<img src="image52_dualaxis.png" width=50%>
- `양쪽 측정값의 축을 동일화` 시키기: 오른쪽 측정값 축 우클릭 > `축 동기화` 클릭
&emsp;&emsp;<img src="image53_dualaxis.png" width=50%><br>
    - 결과: <img src="image54_result.png" width=50%>
    - 고려사항: `두 측정값의 분포 범위를 고려한 후 축 동기화 여부 결정` &rightarrow; 값이 완전 다를 경우 비추천
    - 예) Sales와 Discount의 경우 Sales의 값은 크지만 Discount는 0과 1사이 안에 값이 있기 때문에 축 동기화 시 좋은 시각화가 되지 않을 수 있음
        - 실습 - 그래프 만들기: Sales와 Discount을 행에, Sub Category를 열어 넣어 두 개의 그래프 생성<br>
&emsp;&emsp;<img src="image55_dualaxis.png" width=50%>
        - 위 그래프에 `이중축` > `마크 디자인 조절`<br>
&emsp;&emsp;<img src="image56_dualaxis.png" width=50%>
        - `축 동기화` 적용<br>
&emsp;&emsp;<img src="image57_dualaxis.png" width=50%>
        - 좋은 시각화가 아님 - 값이 잘 보이지 않기에
### + Combined Axis Chart (이중축 아님)
- 만드는 방법:
    - 두번째 측정값을 첫번째 측정값의 축 위로 drag and drop
&emsp;&emsp;<img src="image58_combined.png" width=50%>
&emsp;&emsp;<img src="image59_combined.png" width=50%>
    - 더 잘 구분하기 위해 색상 변경 가능
- 이중축과 Combined Axis Chart의 장단점
    - 이중축의 강점: 마크 카드를 두 개 확보 가능하여 커스터마이징 자요도 &uparrow;
    - Combined Axis Chart의 강점: 이중축을 쓸 수 있는 기회를 남겨둠 (조커를 아낌)
### 2) 강조 포인트 제공하기 위한 이중축 구축
- 그래프 만들기: Sale를 행에, Order Date(연속형 - 월)를 열어 넣어 두 개의 그래프 생성<br>
&emsp;&emsp;<img src="image60_dualaxis.png" width=50%>
    - 특정한 날 (2017년 11월) 강조하기
        - 2017년 11월에 해당하는 별도의 Sales 금액 필요 (두번째 측정값): 왼쪽 바에서 Sales 우클릭 > `만들기` > `계산된 필드` > 다음을 계산식으로 입력:<br> 
&emsp;&emsp;`IIF([Order Date]>= DATE('2017-11-01') AND [Order Date]<= DATE('2017-11-30'), [Sales], NULL)`<br>
&emsp;&emsp;<img src="image61_dualaxis.png" width=50%>
        - 만들어진 `계산 1` (원하면 이름 변경 가능)을 행으로 옮긴 후 확인 
        - 이중축 생성: 계산 1 우클릭 > `이중축` 클릭
        - 축 동기화 설정: 오른쪽 축 우클릭 > `축 동기화` 클릭
        - 다음은 마크 필드에서 원하는 설정하여 디자인 변경 가능<br>
&emsp;&emsp;<img src="image62_dualaxis.png" width=50%>
