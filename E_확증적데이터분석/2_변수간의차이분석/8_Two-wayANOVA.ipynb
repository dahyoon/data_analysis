## 1. (예제 1)
- 태아의 머리 둘레 측정 데이터
- 3명의 태아를 대상으로 4명의 관측자가 측정
- 이를 통해서 초음파로 태아의 머리 둘레측정 데이터가 재현성이 있는지 조사
### 1) 데이터 가져오기

```python
df = read_excel("https://data.hossam.kr/E02/alrtman.xlsx")
df
```

### 2) 탐색적 데이터 분석
#### 측정자에 따른 태아의 머리둘레 확인
- 그림을 보면 태아 3명의 머리둘레는 차이가 있어보이지만 이것이 관측자와 상호작용이 있는 것인지 분석을 통해 알아보고자 함

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.unicode_minus"] = False

plt.figure()
sb.boxplot(data=df, x='태아번호', y='머리둘레', hue='측정자번호')
plt.show()
plt.close()
```

### 3) 추론적 데이터 분석
#### 가설 설정
    - 귀무가설: 태아의 머리 둘레 차이는 측정자와 연관이 없다
    - 대립가설: 태아의 머리 둘레 차이는 측정자와 연관이 있다
#### 정규성, 등분산성, 독립성 검정
> 예제 데이터이므로 충족하다고 가정하고 분석을 진행함

```python
all_test(df['태아번호'], df['측정자번호'], df['머리둘레'])
```

#### 이원분산분석 수행
- `scipy.stats` 패키지는 이원분산분석을 지원하지 않음
    - 머리둘레와 태아
    - 머리둘레와 측정자
    - 머리둘레와 태아 vs 측정자
- 세 가지 상황에 따른 평균에 유의미한 차이가 있는지 분석

```python
formula = '머리둘레 ~ C(태아번호) + C(측정자번호)+ C(태아번호):C(측정자번호)'
lm = ols(formula, df).fit()
anova = anova_lm(lm)
anova
```

#### 결과 해석
- p-value가 0.05보다작으므로 귀무가설을 기각하고 대립가설 채택
- 즉, 측정자(측정장비)와 태아의 머리둘레값에는 연관성이 있다고 할 수 있다.

## 2. (예제 2)
- 한 식물학자는 식물의 성장이 햇빛 노출과 관수 빈도에 의해 영향받는지 여부를 알고 싶어한다.
- 실험은 30개의 씨앗을 심고 햇빛 노출과 물 주기를 다른 조건으로 하여 두 달 동안 자라게 한 후 각 식물의 높이를 인치 단위로 기록하였다.

### 1) 데이터 가져오기

|필드|설명|값|
|--:|--|--|
|water|각 식물에 물을 주는 빈도|매일:daily, 매주:weekly|
|sun|각 식물이 받은 햇빛 노출량|낮음:low, 중간:med, 높음:high|
|height|2개월 후 각 식물의 높이|인치|

```python
df = read_excel('https://data.hossam.kr/E02/plant_growth.xlsx')
df
```

### 2) 탐색적 데이터 분석
#### 요인에 따른 식물의 높이 확인

```python
plt.figure()
sb.boxplot(data=df, x='sun', y='height', hue='water')
plt.show()
plt.close()
```

### 3) 데이터 전처리

```python
df2 = df.copy()
df2['sun'] = df2['sun'].map({'low':0, 'med':1, 'high':2})
df2['water'] = df2['water'].map({'daily':0, 'weekly':1})
df2
```

### 4) 추론적 데이터 분석
#### 정규성, 등분산성, 독립성 확인

```python
all_test(df2['sun'], df2['water'],df2['height'])
```

#### 이원분산분석 수행

```python
model = ols('height ~ C(water) + C(sun) + C(water):C(sun)', data=df).fit()
anova_lm(model)
```

#### 결과 해석
- 물과 태양에 대한 p-값이 모두 `0.05 미만`이므로 두 요인 모두 식물 높이에 통계적으로 `유의미한 영향을 미친다`.
- 그리고 `교호작용 효과(.120667)`에 대한 p-값이 `0.05 이상`이므로 `햇빛노출과 물 주기 사이에 유의한 교호작용 효과가 없음`을 알 수 있다.
- `교호작용(Interaction)`: 한 요인의 효과가 다른 요인의 수준에 의존하는 경우
