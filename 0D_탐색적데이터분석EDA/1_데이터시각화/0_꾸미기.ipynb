# Matplotlib
#### 글꼴 굵기 설정 (fontweight)
1. 숫자로 지정 (100~900):
- 100: 가장 얇은 굵기 (Thin)
- 200: Extra Light
- 300: Light
- 400: Normal (기본값)
- 500: Medium
- 600: Semi Bold (Demi Bold)
- 700: Bold
- 800: Extra Bold
- 900: Black (가장 두꺼운 굵기)
2. 문자로 지정:
- "ultralight": 매우 얇은 글씨
- "light": 얇은 글씨
- "normal": 보통 두께 (기본값)
- "medium": 중간 두께
- "semibold" 또는 "demibold": 약간 굵은 글씨
- "bold": 굵은 글씨
- "heavy" 또는 "extra bold": 매우 굵은 글씨
- "black": 가장 굵은 글씨


# Seaborn
참조:
- https://life-is-egg-study.tistory.com/7(Seaborn 색상 지정)

```python
# 패키지 참조
import sys
import seaborn as sb
from matplotlib import pyplot as plt
from pandas import read_excel
from statsmodels.graphics.mosaicplot import mosaic

# 데이터 가져오기 및 전처리
titanic = read_excel("https://data.hossam.kr/D01/titanic.xlsx", index_col="PassengerId")
df = titanic.astype({"Survived": "category", "Pclass": "category", "Sex": "category", 
                     "Cabin": "category", "Embarked": "category"})
df.head()

# 그래프 전역 설정
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 10
plt.rcParams["figure.figsize"] = (5, 3)
plt.rcParams["axes.unicode_minus"] = False
```

### Seaborn 색상 지정
- 원하는 색상의 색상코드표를 찾은 후, 리스트(list)에 넣고 리스트를 팔레트 형태로 만들어 seaborn에 저장
- `seaborn.set_palette(seaborn.color_palette(리스트))`

```python
sb.set_palette(sb.color_palette(['red', 'blue', 'lightcoral', 'lightblue'])) # hue 색상 지정
sb.scatterplot(data=df, x='Age', y='Fare', hue=df[['Survived', 'Sex']].apply(tuple, axis=1), s=10)
plt.title("연령대, 티켓요금에 따른 생존여부 분포")
plt.grid()
plt.show()
plt.close()
```
