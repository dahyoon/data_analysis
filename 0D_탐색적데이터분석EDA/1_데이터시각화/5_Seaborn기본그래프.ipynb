# Seaborn 기본 그래프

- matplotlib을 기반 python 데이터 시각화 라이브러리
- `다양한 색상 테마`와 `통계용 차트` 등의 기능 추가
- 시각화 할 때 가급적이면 `seaborn` 사용 권장

### 주요 파라미터
| 파라미터 | 설명 |
|---|---|
| `data`ㅤ| 데이터 구조 입력; 일반적으로 데이터프레임 객체 |
| `x`, `y`ㅤ| x, y축의 값을 지정하는 변수<br/>`data`를 미설정시: 리스트 등의 연속형 변수<br/>`data`를 설정시: 데이터프레임의 컬럼이름 |
| `hue` | 데이터를 구분하기 위한 그룹화 변수(범주형 변수);<br/>`data`를 설정시 데이터프레임의 컬럼이름 |
| `size`ㅤ| 너비가 다른 선을 생성하는 그룹화 변수 (마커의 경우, 마커의 크기가 다르게 나타남) |
| `style`ㅤ| 다른 스타일의 마커 또는 선을 생성하는 그룹화 변수 |
| `palette`ㅤ| 색상 체계 선택:<br/> deep, muted, pastel, bright, dark, colorblind, Spectral |

### 1. 기본 준비

#### 1) 패키지 참조
- seaborn 패키지 설치 필요

```python
from matplotlib import pyplot as plt
from pandas import read_excel, DataFrame
import seaborn as sb
```

#### 2) 그래프 전역 설정

```python
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.unicode_minus'] = False
```

### 2. 기본 그래프
- pie 그래프는 제공되지 않음

#### 1) 선 그래프

```python
# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)
sb.lineplot(x=[-20, -10, 0, 10, 20], y=[1, 3, 5, 7, 9], label = 'data1')
sb.lineplot(x=[-20, -10, 0, 10, 20], y=[1, 2, 3, 4, 5], label = 'data2')

# 그 밖의 옵션은 matplotlib와 동일
plt.title('한글 제목')
plt.xlabel('엑스축 라벨')
plt.ylabel('와이축 라벨')
plt.legend()
plt.grid()
plt.show()
plt.close()
```

#### 2) 세로 막대 그래프 
- .barplot(x = 리스트, y = 리스트)

```python
x = ["서울", "부산", "제주"]
y = [20, 30, 10]

sb.barplot(x = x, y = y)

plt.title("Bar Chart")
plt.xlabel("지역")
plt.ylabel("빈도수")
plt.grid()
plt.show()
plt.close()
```

#### 3) 가로 막대 그래프
- x축과 y축을 바꿔서 설정! `.barplot(x=y, y=x)`
- 기본 제공 팔레트: deep, muted, pastel, bright, dark, colorblind, Spectral

```python
x = ['서울', '부산', '제주']
y = [20, 30, 10]

sb.barplot(x = y, y = x, label = ['data1', 'data2', 'data2'], palette='muted')

plt.grid()
plt.title("Bar Chart")
plt.xlabel('빈도수')
plt.ylabel('지역')
plt.legend()
plt.show()
plt.close()
```

#### 4) 산점도 그래프

```python
# 온도와 아이스크림 판매 수량
tmp = [23, 25, 26, 27, 28, 29, 30, 31, 33]
qty = [21, 23, 25, 28, 33, 35, 36, 32, 39]

sb.scatterplot(x=tmp, y=qty, label = '판매수량')

plt.title("Scatter Plot")
plt.grid()
plt.xlabel('판매수량')
plt.ylabel('빈도수')
plt.show()
plt.close()
```

### 3. 데이터 프레임을 통한 그래프

#### 1) 데이터 준비

> 이전 단원과 동일한 데이터

```python
발생건수 = [214171, 213745, 211662, 215822, 231990, 226878, 221711, 223656, 215354,
         223552, 232035, 220917, 216335, 217148]
사망자수 = [6376, 6327, 6166, 5870, 5838, 5505, 5229, 5392, 5092,
         4762, 4621, 4292, 4185, 3781]
부상자수 = [342233, 340229, 335906, 338962, 361875, 352458, 341391, 344565, 328711,
         337497, 350400, 331720, 322829, 323037]

df = DataFrame({'년도': range(2005, 2019),
                '발생건수': 발생건수,
                '사망자수': 사망자수,
                '부상자수': 부상자수}).set_index('년도')
df
```

#### 2) 막대 그래프
- .barplot()

```python
# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)
sb.barplot(data = df, x = df.index, y = '사망자수')

plt.title ('교통사고 사망자수')
plt.xlabel('년도')
plt.ylabel('사망자수')
plt.grid()
plt.show()
plt.close()
```

#### 3) 산점도 그래프
- .scatterplot()

```python
# 그래프 그리기 (서브플롯을 사용할 경우 ax 파라미터 사용 가능)
sb.scatterplot(data = df, x = '발생건수', y = '사망자수', label = '사망자수', color = 'pink')

plt.title('교통사고 발생현황')
plt.xlabel('발생건수')
plt.ylabel('사망자수')
plt.legend()
plt.grid()
plt.show()
plt.close()
```

#### 4) 추세선을 포함한 산점도 그래프
- .regplot()
- 추세선 주변에 보이는 구간은 95% 신뢰구간임
- `fit_reg=False` 추가시 추세선이 표시되지 않으며 .scatterplot() 메서드와 동일한 결과를 보여준다 (기본값 = True)
- `ci`파라미터는 신뢰수준(p-value)을 설정 (기본값 = 95)
- `line_kws` 파라미터는 선의 색상
- `scatter_kws` 파라미터는 점의 색상

```python
# 추세선을 함께 보여주는 산점도 그래프 그리기
sb.regplot(data = df, x = '발생건수', y = '사망자수', label = '사망자수',
           ci = 95,                              # 신뢰수준
           line_kws = {'color': 'red'},          # 선 색상
           scatter_kws = {'color': 'lightblue'}) # 점 색상

plt.title('교통사고 발생현황')
plt.xlabel('발생건수')
plt.ylabel('사망자수')
plt.legend()
plt.grid()
plt.show()
plt.close()
```

#### 5) 서브플롯에 기반한 선 그래프
- 서브플롯을 사용하지 않을 경우 `sb.lineplot()` 함수에서 `ax` 파라미터 제거

```python
fig, ax1 = plt.subplots(1, 1)
ax2 = ax1.twinx()

# 그래프 그리기
sb.lineplot(data = df, x = df.index, y = '발생건수', color = 'pink', ax = ax1)
sb.lineplot(data = df, x = df.index, y = '부상자수', color = 'lightblue', ax = ax2)

plt.title('교통사고 발생현황')
plt.xlabel('년도')
plt.ylabel('교통사고')
plt.grid()
plt.show()
plt.close()
```
