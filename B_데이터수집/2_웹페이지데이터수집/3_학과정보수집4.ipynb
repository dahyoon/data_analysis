## 숙명여대 학과 정보 수집 (20240106_다윤 연습)

```python
# 필요한 모듈 참조
import requests
    # request로 갖고 온 데이터값이 JSON형태이면 dictionary로 변환하면 된다
    # request로 갖고 온 데이터값이 HTML + CSS 이면 BeautifulSoup라는 클래스를 통해 HTML + CSS 안에 있는 내용들을 추출해준다
from bs4 import BeautifulSoup
from pandas import DataFrame
```

```python
# 접속개체 생성
session = requests.Session()

# 접속객체에 부가정보 (header) 삽입하기
session.headers.update({
    "Referer": "",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
})
```

```python
# 대학 이름 리스트 정의 --> URL을 분석해서 얻은 값
clist = ['liberal-arts01', 'liberal-arts02','liberal-arts03', 'liberal-arts04',
         'liberal-arts05', 'liberal-arts06', 'liberal-arts07', 'liberal-arts08',
         'liberal-arts09', 'liberal-arts10', 
         'science01', 'science02', 'science03', 'science04', 'science05', 'science06', 
         'engineering01', 'engineering02', 'engineering03', 'engineering04',
         'engineering05','engineering06', 'engineering07', 'engineering08', 
         'life-sciences01', 'life-sciences02', 'life-sciences03', 'life-sciences04',
         'social-sciences01', 'social-sciences02', 'social-sciences03', 'social-sciences04',
         'social-sciences05', 
         'law', 
         'gyeongsang01','gyeongsang02',
         'music01', 'music02', 'music03', 'music04',
         'pharmacy', 
         'arts01', 'arts02', 'arts03', 'arts04', 'arts05',
         'general-education01', 'general-education02',
         'global-service01','global-service02',
         'english01','english02',
         'media01']
```

#### 응답결과로부터 데이터 추출하기

```python
# 수집한 정보를 저장할 빈 리스트
학과목록_리스트 = []

# 대학 수 만큼 반복
for c in clist:
    # 수집할 컨텐츠가 있는 웹 페이지의 주소
    if c == 'media01':
        url = "https://www.sookmyung.ac.kr/kr/intro/%s.do" % c
    else:
        url = "https://www.sookmyung.ac.kr/kr/university-graduate/%s.do" % c

    # 생성한 접속객체를 활용하여 API에 접속
    r = session.get(url)

    # 접속에 실패한 경우
    if r.status_code != 200:
        # 에러코드와 에러메시지 출력
        msg = "[%d Error] %s 에러가 발생함" % (r.status_code, r.reason)
        # 에러를 강제로 생성시킴
        raise Exception (msg)

    r.encoding = "utf-8"
    soup = BeautifulSoup(r.text)
    
    content_wrap = soup.select(".content-wrap")

    for item in content_wrap:
        # print(item)

        # 대학 계열 이름
        h3El = item.select("h3")
        name = h3El[0].text.strip()
        print('대학 계열 이름:', name)
    
        # 학과 이름
        titleEl = item.select("h4")
        title = titleEl[0].text.strip()
        print('학과 이름:', title)

        # 학과 홈페이지
        homepageEl = item.select(".home")
        if homepageEl[0].text.strip() == '홈페이지':
            homepage = homepageEl[0].attrs['href']
        print('학과 홈페이지:', homepage)

        # 학과 소개 pdf 
        if item.select(".download"):
            infoEl = item.select(".download")
            info = infoEl[0].attrs['href']
        else:
            info = homepageEl[1].attrs['href']
        print('학과 소개 pdf/페이지:',info)

        # 간단한 소개
        short_introEl = item.select(".con-p")
        short_intro = short_introEl[0].text.strip()
        print('간단한 소개:', short_intro)
        print("-" * 50)

        # 수집한 값들을 딕셔너리로 묶은 후 리스트에 추가
        college_dict = {'대학계열이름': name,
                        '학과이름': title,
                        '학과 홈페이지': homepage,
                        '학과 소개 pdf/페이지': info,
                        '간단한 소개': short_intro}
        학과목록_리스트.append(college_dict)
학과목록_리스트
df = DataFrame(학과목록_리스트)
df.index += 1 # index를 1부터 시작하기
df.to_excel('3_학과정보수집4.xlsx')
df
```
