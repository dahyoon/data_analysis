## 5. hwp 파일 (*.hwp)
- `olefile` 패키지 설치 필요
    - `pip3 install --upgrade olefile` 라고 터미널에 입력하여 다운로드

```python
import olefile 
f = olefile.OleFileIO('/Users/dayoonz/Documents/8 데이터분석/프로젝트4_아동학대/데이터/datasetDetail.hwp') # HWP 파일 열기
encoded_text = f.openstream('PrvText').read() # PrvText 스트림의 내용 꺼내기
decoded_text = encoded_text.decode('UTF-16') # 유니코드를 UTF-16으로 디코딩
print(decoded_text)
```
