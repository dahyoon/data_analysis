# 상관분석
- 명목형 (순서가 없는 범주를 가진 경우- 성별, 혈액형 등) 분석에는 적용 불가

# 피어슨 상관분석
- 피어슨 상관계수는 두 변수 간의 `선형 상관관계`를 측정
- `연속적인 데이터`에서 사용되며, 두 변수가 서로 선형적으로 관련되어 있는 경우 유용
- 두 변수가 정규 분포를 따르고, `이상치(outliers)가 적은 경우` 사용 가능
- 값의 범위: `-1 에서 +1`
    - 1에 가까울수록 양의 선형 상관관계
    - -1에 가까울수록 음의 선형 상관관계
    - 0에 가까울수록 상관관계가 없음

## 1. 작업 준비
### 패키지 가져오기

```python
import numpy as np
from pandas import read_excel
from scipy import stats

import sys
import seaborn as sb
from matplotlib import pyplot as plt
```

## (예제 1) 언어영역 점수와 수리영역 점수의 상관관계
### 1. 데이터 가져오기

```python
df = read_excel("https://data.hossam.kr/E03/point.xlsx")
df
```

### 2. 탐색적 데이터 분석
#### 1) 산점도 그래프를 통한 상관정도 확인

```python
plt.rcParams["font.family"] = 'AppleGothic' if sys.platform == 'darwin' else 'Malgun Gothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.unicode_minus"] = False

# sb.scatterplot(data=df, x='언어영역', y='수리영역')
plt.title('언어영역 & 수리영역 상관정도')
sb.regplot(data=df, x='언어영역', y='수리영역')
plt.grid()
plt.show()
plt.close()
```

### 3. 공분산
- `numpy.cov(변수값컬럼1, 변수값컬럼2)[0, 1]`
- 공분산은 상관관계의 상승 혹은 하강하는 경향을 이해할 수는 있으나 2개 변수의 측정 단위의 크기에 따라 값이 달라지므로` 절대적 정도를 파악하기에는 한계`가 있다.
- 결과 값: `양수 = 정비례`, `음수 = 반비례`
> 그래서 잘 안 쓴다.

```python
np.cov(df['언어영역'], df['수리영역'])[0, 1]
```

### 4. 상관분석
#### 1) numpy를 통해 상관계수 구하기
- `numpy.corrcoef(변수값컬럼1, 변수값컬럼2)`
- `p-value`는 알 수 없기에 `분석결과 검정 불가`
- 기본적으로 상관행렬 표시

```python
np.corrcoef(df['언어영역'], df['수리영역'])
```

- 상관계수를 얻기 위해선 `0,1` 혹은 `1,0`번째 항목을 취함

```python
np.corrcoef(df['언어영역'], df['수리영역'])[0, 1]
```

#### 2) 피어슨 상관계수와 `p-value` 구하기
- `r, p = stats.pearsonr(변수값컬럼1, 변수값컬럼2)`
    - `r = 상관계수`
    - `p = p-value`

```python
r, p = stats.pearsonr(df['언어영역'], df['수리영역'])
"상관계수: %f, p-value: %f" % (r, p)
```

##### 결과 해석
- `p-value`가 유의수준 `0.05보다 작으므로` 귀무가설을 기각하고 대립가설을 채택한다. 그러므로 `언어영역 점수와 수리영역 점수는 상관이 있다`고 볼 수 있다.
#### 3) 데이터프레임 객체를 통해 상관계수 구하기
- `데이터프레임.corr(method='pearson')`
- numpy를 통해 구한 결과와 동일
- 이와 같이 표시되는 형태를 `상관행렬`이라고 함

```python
corr_df = df.corr(method='pearson')
corr_df
```

#### 4) 상관분석 결과 시각화
- `sb.heatmap(데이터프레임, annot = True)`

```python
plt.rcParams['figure.figsize']=(5,3)
sb.heatmap(corr_df,
           annot = True,        # 실제 값 화면에 나타내기
           cmap = 'Reds',      # 색상
           vmin = -1, vmax = 1, # 컬러차트 영역 -1 ~ +1
           linewidths = 0.5     # 구분선 넓이
)
plt.show()
plt.close()
```

## (예제 2) mtcars 데이터를 통해 실린더 수와 연비의 상관관계 분석
- R 패키지인 ggplot2에 내장된 샘플 데이터로 자동차 32종의 11개 속성에 대한 정보를 담고 있는 데이터

|필드|설명|
|--:|--|
|mpg|연료 효율성 (마일 당 갤런)|
|cyl|실린더 수|
|disp|배기량 (cubic inches)|
|hp|마력 (horsepower)|
|drat|후륜축 비율 (rear axle ratio)|
|wt|차량 무게 (1000 파운드)|
|qsec|1/4 마일 달성 시간 (초)|
|vs|V/S (V 엔진 형태인지, S는 직렬 엔진 형태인지를 나타냄)|
|am|변속기 형태 (0: 자동, 1: 수동)|
|gear|전진 기어 수 (전송 기어)|
|carb|기화기 개수 (carburetors)|

### 1. 데이터 가져오기

```python
df = read_excel("https://data.hossam.kr/E03/mtcars.xlsx", index_col='name')
df
```

### 2. 실린더 수에 따른 연비에 대한 `산점도 그래프`

```python
plt.rcParams["figure.figsize"] = (10, 5)

sb.regplot(data=df, x='cyl', y='mpg')
plt.grid()
plt.show()
plt.close()
```

- 실린더 수가 증가함에 따라 연비가 낮아지는 경향이 있는 것으로 보임

### 3. 상관분석

```python
plt.rcParams["figure.figsize"] = (5,3)

corr_df = df.filter(['cyl', 'mpg']).corr(method='pearson')
sb.heatmap(corr_df,
           annot = True,            # 실제 값 화면에 나타내기
           cmap = 'Blues',          # 색상
           vmin = -1, vmax = 1,    # 컬러 차트 영역 -1 ~ +1
           linewidths =3)           # 구분선 넓이
plt.show()
plt.close()
```

##### 분석 결과
- 상관계수가 `-0.85`로 나타나 실린더 수와 연비는 `강한 음의 상관관계가 있는 것`으로 나타난다.
##### 분석 결과 검정

```python
stats.pearsonr(df['cyl'], df['mpg'])
```

##### 결과 보고
- 32종의 자동차를 대상으로 실린더 수에 따른 연비를 조사한 결과 `실린더 수와 연비는 강한 음의 상관관계를 갖는 것`으로 나타났다. `p-value`가 유의수준 `0.05보다 작으므로` 이는 통계적으로 유의하다.
