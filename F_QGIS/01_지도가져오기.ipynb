### 1) Shape 파일의 구성
- 압축을 풀면 4개의 파일(*.dbf, *.prj, *.shp, *.shx)로 구성되어있는데, 이 4개의 파일이 한 세트이므로 반드시 동일 폴더 안에 포함되어 있어야 한다

    |포멧|설명|
    |--|--|
    |`*.dbf`|각 도형의 속성 정보가 저장되어 있는 파일|
    |`*.prj`|좌표정보가 저장되어 있는 파일|
    |`*.shp`|shape format. 도형파일(백터형식)|
    |`*.shx`|shape index format. 도형의 위치, 방향 정보가 저장되어 있는 파일|

### 2) GeoJson
- `*.shp`을 JSON 형식으로 내보낸 형태
- 주로 웹 프로그램에서 활용하기 위한 용도

## 2. 배경지도
- shape 뒤에 배경으로 배치되는 지도
- 대체로 이미지, 구글맵, 네이버 지도 등을 배경으로 설정할 수 있다
- 구글 지도 경우 자세히 보면 우리나라가 배포하는 지도와 정확히 일치하지 않는다
    - 이유: 대한민국 법률상 우리나라 지도 데이터를 해외로 내보내는 것이 불법이기 때문에 구글이 우리나라 지도를 정확히 표현 못한다 (전쟁국가이기 때문에)
- 정확한 한국 지도를 사용하려면 NAVER나 Kakao 지도 권장
### 1) 이미지 타일지도
- 하나의 이미지가 아닌 복수개의 지도 이미지를 사용자가 원하는 영역에 축척별로 다운로드 받아 사용하는 것
### 2) 이미지 타일지도 실습
#### step1. 새 연결 생성
- 탐색기에서 `XYZ Tiles` 항목을 우클릭하고 'New Connection (새 연결)' 선택 > Connection Details: Name: Google Satellite, URL: https://mt1.google.com/vt/lyrs=m&hl=ko&x={x}&y={y}&z={z} 입력
    - URL에 넣을 수 있는 것들:
        - Roadmap (일반도) - https://mt1.google.com/vt/lyrs=m&hl=ko&x={x}&y={y}&z={z}
        - Terrain - https://mt1.google.com/vt/lyrs=p&hl=ko&x={x}&y={y}&z={z}
        - Altered roadmap - https://mt1.google.com/vt/lyrs=r&hl=ko&x={x}&y={y}&z={z}
        - Satellite only - https://mt1.google.com/vt/lyrs=s&hl=ko&x={x}&y={y}&z={z}
        - Terrain only - https://mt1.google.com/vt/lyrs=t&hl=ko&x={x}&y={y}&z={z}
        - Hybrid - https://mt1.google.com/vt/lyrs=y&hl=ko&x={x}&y={y}&z={z}
