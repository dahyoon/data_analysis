# 문제 1
- 아래의 '콩의 수분 함유량에 따른 강도를 조사한 데이터'를 분석하라

https://data.hossam.kr/E04/beans.xlsx

```python
from pandas import read_excel
from statsmodels.formula.api import ols
from matplotlib import pyplot as plt
import seaborn as sb
import os, sys

sys.path.append(os.path.dirname(os.path.dirname(os.getcwd())))
from helper import pearson_r
```

```python
df = read_excel("https://data.hossam.kr/E04/beans.xlsx")

# 탐색적 데이터 분석 - 산점도 그래프
print("1. 탐색적 데이터 분석: 산점도 그래프를 통해 상관관계 추론")
plt.rcParams["font.family"] = 'AppleGothic'
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (5, 2)
plt.rcParams["axes.unicode_minus"] = False

sb.regplot(data=df, x='수분함유량', y='강도')
plt.grid()
plt.show()
plt.close()

# 추론 통계 - 1) 상관분석
print("2. 추론 통계 \n 1) 상관분석: ")
print(pearson_r(df))
print("상관분석 결과 p-value가 0.05보다 작으므로 콩의 수분함유량과 강도 간 양의 상관관계가 통계적으로 유의함을 알 수 있다.")

# 추론 통계 - 2) 회귀 분석
print(" 2) 회귀 분석: ")
model = ols('강도 ~ 수분함유량', data = df)
fit = model.fit()
print(fit.summary())
print("  1-모형적합도 보고:")
print("dist에 대하여 speed로 예측하는 회귀분석을 실시한 결과, 이 회귀모형은 통계적으로 유의미하였다.")
print("F(1, 8) = 29.85, p < 0.05")
print("  2-독립변수에 대한 보고:")
print("수분함유량의 회귀계수는 0.7123로, 강도에 대하여 유의미한 예측변인인 것으로 나타났다.")
print("t(8)= 5.463, p < 0.05")
```
